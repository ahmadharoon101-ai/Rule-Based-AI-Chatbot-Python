```python
import time
import sys

# в”Җв”Җ ANSI COLOR CODES в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ
class Color:
    RESET       = "\033[0m"
    BOLD        = "\033[1m"
    DIM         = "\033[2m"

    CYAN        = "\033[96m"
    DARK_CYAN   = "\033[36m"
    GREEN       = "\033[92m"
    YELLOW      = "\033[93m"
    RED         = "\033[91m"
    BLUE        = "\033[94m"
    MAGENTA     = "\033[95m"
    WHITE       = "\033[97m"
    GRAY        = "\033[90m"
    BG_DARK     = "\033[40m"
    BG_BLUE     = "\033[44m"
    BG_CYAN     = "\033[46m"

# в”Җв”Җ SYMBOLS в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ
class Icon:
    BOT      = "рҹӨ–"
    USER     = "рҹ‘Ө"
    SUCCESS  = "вң…"
    ERROR    = "вқҢ"
    ARROW    = "вһӨ"
    DOT      = "в—Ҹ"
    LINE     = "в”Җ"
    SPARK    = "вҡЎ"
    BRAIN    = "рҹ§ "
    ROCKET   = "рҹҡҖ"
    SHIELD   = "рҹӣЎпёҸ"
    GEAR     = "вҡҷпёҸ"
    CHIP     = "рҹ’Ў"

# в”Җв”Җ TYPING ANIMATION в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ
def typing_effect(text, delay=0.018):
    """Simulate BotX typing character by character."""
    for char in text:
        sys.stdout.write(char)
        sys.stdout.flush()
        time.sleep(delay)
    print()

def thinking_animation():
    """Show a thinking animation while BotX processes."""
    sys.stdout.write(f"  {Color.GRAY}BotX is thinking")
    sys.stdout.flush()
    for _ in range(3):
        time.sleep(0.25)
        sys.stdout.write(f"{Color.CYAN}.")
        sys.stdout.flush()
    time.sleep(0.2)
    # Clear the line
    sys.stdout.write("\r" + " " * 30 + "\r")
    sys.stdout.flush()

# в”Җв”Җ BANNER в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ
def print_banner():
    banner = f"""
{Color.CYAN}{Color.BOLD}
  в•”в•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•—
  в•‘                                                      в•‘
  в•‘   {Color.WHITE}в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•—  в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•— в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•—в–Ҳв–Ҳв•—  в–Ҳв–Ҳв•—{Color.CYAN}              в•‘
  в•‘   {Color.WHITE}в–Ҳв–Ҳв•”в•җв•җв–Ҳв–Ҳв•—в–Ҳв–Ҳв•”в•җв•җв•җв–Ҳв–Ҳв•—в•ҡв•җв•җв–Ҳв–Ҳв•”в•җв•җв•қв•ҡв–Ҳв–Ҳв•—в–Ҳв–Ҳв•”в•қ{Color.CYAN}              в•‘
  в•‘   {Color.WHITE}в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•”в•қв–Ҳв–Ҳв•‘   в–Ҳв–Ҳв•‘   в–Ҳв–Ҳв•‘    в•ҡв–Ҳв–Ҳв–Ҳв•”в•қ {Color.CYAN}              в•‘
  в•‘   {Color.WHITE}в–Ҳв–Ҳв•”в•җв•җв–Ҳв–Ҳв•—в–Ҳв–Ҳв•‘   в–Ҳв–Ҳв•‘   в–Ҳв–Ҳв•‘    в–Ҳв–Ҳв•”в–Ҳв–Ҳв•— {Color.CYAN}              в•‘
  в•‘   {Color.WHITE}в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•”в•қв•ҡв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•”в•қ   в–Ҳв–Ҳв•‘   в–Ҳв–Ҳв•”в•қ в–Ҳв–Ҳв•—{Color.CYAN}              в•‘
  в•‘   {Color.WHITE}в•ҡв•җв•җв•җв•җв•җв•қ  в•ҡв•җв•җв•җв•җв•җв•қ    в•ҡв•җв•қ   в•ҡв•җв•қ  в•ҡв•җв•қ{Color.CYAN}              в•‘
  в•‘                                                      в•‘
  в•‘   {Color.DARK_CYAN}Rule-Based AI Chatbot  {Color.YELLOW}в–ё Project 1{Color.CYAN}            в•‘
  в•‘   {Color.GRAY}DecodeLabs Industrial Training Kit В· Batch 2026{Color.CYAN}  в•‘
  в•‘   {Color.GRAY}Architecture : IPO Model В· O(1) Dictionary Lookup{Color.CYAN}в•‘
  в•‘                                                      в•‘
  в•ҡв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•җв•қ
{Color.RESET}"""
    print(banner)

# в”Җв”Җ DIVIDER в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ
def divider(char="в”Җ", color=Color.DARK_CYAN, width=56):
    print(f"  {color}{char * width}{Color.RESET}")

def section(title):
    pad = (54 - len(title)) // 2
    print(f"\n  {Color.CYAN}в•”{'в•җ' * 54}в•—{Color.RESET}")
    print(f"  {Color.CYAN}в•‘{' ' * pad}{Color.WHITE}{Color.BOLD}{title}{Color.RESET}{Color.CYAN}{' ' * (54 - pad - len(title))}в•‘{Color.RESET}")
    print(f"  {Color.CYAN}в•ҡ{'в•җ' * 54}в•қ{Color.RESET}\n")

# в”Җв”Җ STARTUP SEQUENCE в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ
def startup_sequence():
    print_banner()
    time.sleep(0.3)

    print(f"  {Color.GRAY}Initializing BotX Engine...{Color.RESET}")
    checks = [
        ("Knowledge Base loaded",    "25 intents registered"),
        ("Sanitization module",       "lower() + strip() active"),
        ("Infinite loop engine",      "while True : READY"),
        ("Exit strategy",             "kill command armed"),
        ("Fallback handler",          "default response set"),
    ]
    for label, detail in checks:
        time.sleep(0.2)
        print(f"  {Color.GREEN}  {Icon.SUCCESS} {Color.WHITE}{label:<28}{Color.GRAY} {detail}{Color.RESET}")

    time.sleep(0.3)
    divider()
    print(f"\n  {Icon.BOT}  {Color.CYAN}{Color.BOLD}BotX is ONLINE {Color.GREEN}{Icon.DOT}{Color.RESET}")
    print(f"  {Color.GRAY}Type {Color.YELLOW}'help'{Color.GRAY} to see commands В· Type {Color.RED}'exit'{Color.GRAY} to quit{Color.RESET}\n")
    divider()
    print()

# в”Җв”Җ KNOWLEDGE BASE в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ
responses = {
    "hello"                       : "Hello! Welcome to DecodeLabs AI Assistant. How can I help you today?",
    "hi"                          : "Hi there! I'm your rule-based AI assistant. Ask me anything!",
    "hey"                         : "Hey! Good to see you. What's on your mind?",
    "what is your name"           : "I'm BotX вҖ” a rule-based AI built at DecodeLabs, Batch 2026.",
    "who are you"                 : "I am BotX, an intelligent chatbot powered by deterministic logic вҖ” no hallucinations, pure precision.",
    "who made you"                : "I was engineered by an AI intern at DecodeLabs, Batch 2026.",
    "how are you"                 : "All systems running at 100% efficiency вҖ” no bugs detected! How about you?",
    "are you okay"                : "All systems operational. Thanks for checking in!",
    "what can you do"             : "I answer predefined questions, explain AI concepts, share facts, and hold structured conversations.",
    "help"                        : "Commands вҶ’ 'hello' В· 'what is ai' В· 'tell me a joke' В· 'what is python' В· 'who made you' В· 'exit'",
    "what is ai"                  : "AI is the simulation of human intelligence by machines using logic, data, and algorithms.",
    "what is machine learning"    : "ML is a subset of AI where systems learn patterns from data вҖ” without being explicitly programmed.",
    "what is a chatbot"           : "A chatbot simulates conversation. I'm rule-based вҖ” strict deterministic logic, zero randomness.",
    "what is python"              : "Python is a high-level language widely used in AI, data science, automation, and web development.",
    "what is a dictionary"        : "In Python, a dictionary is a key-value data structure with O(1) lookup вҖ” the engine behind my responses!",
    "what is decodelabs"          : "DecodeLabs is an AI & tech training company in Greater Lucknow, India вҖ” building the next generation of engineers.",
    "tell me a joke"              : "Why do programmers prefer dark mode? Because light attracts bugs! рҹҳ„",
    "tell me a fact"              : "The first chatbot ELIZA was created in 1966 at MIT вҖ” also rule-based, just like me!",
    "what is the meaning of life" : "42. At least, that's what the algorithm outputs.",
    "what is ipo model"           : "IPO = Input вҶ’ Process вҶ’ Output. It's the foundational blueprint for all AI systems, including me.",
    "what is o1 complexity"       : "O(1) means constant time вҖ” no matter how many rules exist, dictionary lookup takes the same time!",
    "bye"                         : "Goodbye! Keep building great things. рҹҡҖ",
    "goodbye"                     : "See you next time! DecodeLabs is proud of your progress.",
    "thanks"                      : "You're welcome! That's what I'm here for.",
    "thank you"                   : "Anytime! Happy to help.",
}

# в”Җв”Җ SANITIZATION в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ
def get_response(user_input):
    clean_input = user_input.lower().strip()     # Phase 1: Sanitize
    if clean_input in ("exit", "quit"):
        return "__EXIT__"
    if clean_input == "":
        return "__EMPTY__"
    return responses.get(clean_input, None)      # Phase 2: O(1) Lookup

# в”Җв”Җ PRINT USER MESSAGE в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ
def print_user(text):
    print(f"\n  {Color.BLUE}{Icon.USER}  {Color.WHITE}{Color.BOLD}You{Color.RESET}  {Color.GRAY}в–ё{Color.RESET}")
    print(f"  {Color.BLUE}в”Ӯ{Color.RESET}  {Color.WHITE}{text}{Color.RESET}")

# в”Җв”Җ PRINT BOT MESSAGE в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ
def print_bot(text, color=Color.CYAN):
    print(f"\n  {Icon.BOT}  {color}{Color.BOLD}BotX{Color.RESET}  {Color.GRAY}в–ё{Color.RESET}")
    sys.stdout.write(f"  {Color.DARK_CYAN}в”Ӯ{Color.RESET}  {color}")
    sys.stdout.flush()
    typing_effect(text)
    print(Color.RESET, end="")

# в”Җв”Җ MAIN CHATBOT LOOP в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ
def run_chatbot():
    startup_sequence()

    msg_count = 0

    while True:
        # в”Җв”Җ INPUT в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ
        try:
            sys.stdout.write(f"  {Color.YELLOW}{Color.BOLD}You в–ё {Color.RESET}{Color.WHITE}")
            sys.stdout.flush()
            user_input = input()
            print(Color.RESET, end="")
        except (EOFError, KeyboardInterrupt):
            print()
            print_bot("Session interrupted. Goodbye! рҹ‘Ӣ", Color.RED)
            break

        print_user(user_input)

        # в”Җв”Җ PROCESS в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ
        thinking_animation()
        reply = get_response(user_input)
        msg_count += 1

        # в”Җв”Җ OUTPUT в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ
        if reply == "__EXIT__":
            divider("в•җ", Color.CYAN)
            print_bot("Session terminated. Goodbye, Engineer! рҹ‘Ӣ", Color.GREEN)
            print(f"\n  {Color.GRAY}Messages exchanged : {Color.CYAN}{msg_count}{Color.RESET}")
            print(f"  {Color.GRAY}Status             : {Color.GREEN}Completed вң…{Color.RESET}")
            divider("в•җ", Color.CYAN)
            print(f"\n  {Color.YELLOW}{Color.BOLD}DecodeLabs В· Batch 2026 В· Project 1 Complete{Color.RESET}\n")
            break

        elif reply == "__EMPTY__":
            print_bot("Please type something! Try 'help' for available commands.", Color.YELLOW)

        elif reply is None:
            print_bot(
                f"I don't understand that yet. {Icon.CHIP} Type 'help' to see what I can do!",
                Color.YELLOW
            )

        else:
            print_bot(reply, Color.CYAN)

        print(f"\n  {Color.GRAY}{'в”Җ' * 54}{Color.RESET}")

# в”Җв”Җ ENTRY POINT в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ
if __name__ == "__main__":
    run_chatbot()
```
